# Problem 2

```python
def Gaussian(lamb, n):
    #(1/np.sqrt(2*np.pi*n)) * np.exp(-0.5 * ((n - lamb)/np.sqrt(lamb))**2)
    return (1/np.sqrt(2*np.pi*n)) * np.exp(-0.5 * ((n - lamb)/np.sqrt(lamb))**2)

def logPoisson(lamb, n):
    # using log of the poisson to avoid taking large factorials
    # with stirling's approximation
    '''added a normalization term - check that this makes sense'''
    return np.log((1/np.sqrt(2*np.pi*n))) + n*np.log(lamb/n) + (n - lamb)
```

```python
lambs = np.linspace(1, 1000, 1001)

# 3 sigma
lamb3 = 0
for lamb in lambs:
    sig = np.sqrt(lamb)
    n3 = lamb + 3*sig
    G = Gaussian(lamb, n3)
    P = np.exp(logPoisson(lamb, n3))
    ratio = P/G
    if ratio <= 2:
        lamb3 = lamb
        print('n =', n3, 'Gives P(n)/G(n) ~', round(ratio, 3), r'at 3sigma')
        break
        
# 5 sigma
lamb5 = 0
for lamb in lambs:
    sig = np.sqrt(lamb)
    n5 = lamb + 5*sig
    G = Gaussian(lamb, n5)
    P = np.exp(logPoisson(lamb, n5))
    ratio = P/G
    if ratio <= 2:
        lamb5 = lamb
        print('n =', n5, 'Gives P(n)/G(n) ~', round(ratio, 3), r'at 5sigma')
        break     
```

```python
npts = int(1e3)
sig3, sig5 = np.sqrt(lamb3), np.sqrt(lamb5)
nmin, nmax = 1, lamb3 + 6*sig3
nn = np.linspace(nmin, nmax, npts)
Gfig, Pfig = [], []
for n in nn:
    Gfig.append(Gaussian(lamb3, n))
    Pfig.append(np.exp(logPoisson(lamb3, n)))

plt.plot(nn, Gfig, label='Gaussian approx')
plt.plot(nn, Pfig, label='Poisson')
plt.axvspan(lamb3 - 3*sig3, lamb3 + 3*sig3, alpha=0.1, color='purple', label=r'$3\sigma$')
plt.legend()
plt.xlabel('n')
plt.ylabel('Prob Density')
plt.show()

nmin, nmax = 600, lamb5 + 6*sig5
nn = np.linspace(nmin, nmax, npts)
Gfig, Pfig = [], []
for n in nn:
    Gfig.append(Gaussian(lamb5, n))
    Pfig.append(np.exp(logPoisson(lamb5, n)))

plt.plot(nn, Gfig, label='Gaussian approx')
plt.plot(nn, Pfig, label='Poisson')
plt.axvspan(lamb5 - 5*sig5, lamb5 + 5*sig5, alpha=0.1, color='purple', label=r'$5\sigma$')
plt.legend()
plt.xlabel('n')
plt.ylabel('Prob Density')
plt.show()
```

```python
plt.plot(x, pred_poly, label='fit')
plt.plot(x, signal, label='signal')
plt.plot(x, data, '.', label='noisy data')
plt.legend()
plt.show()
```

```python
'''Fitting 10000 chunks of the data'''
times = 10000
npts = 51
order = 4
npts = 51
x = np.linspace(-5, 5, npts)
A=np.polynomial.polynomial.polyvander(x, order)

amps = []
vari = []

for t in range(times):
    noise = np.random.randn(npts)
    signal = np.exp(-x**2/2)
    data = noise + signal

    N = np.diag(noise**2)
    Ninv = np.linalg.inv(N)

    lhs=A.T@Ninv@A
    rhs=A.T@Ninv@data
    var = np.linalg.inv(lhs)
    pp=var@rhs
    amps.append(pp[0])
    vari.append(var[0,0])
```

```python
# estimating the variance weighted average of the fitted amplitude of a single chunk
bias = amps[0] - 1
significance = bias / np.mean(noise_chunk)
print('Amp =', round(amps[0], 3), 'with uncertainty', round(np.sqrt(vari[0]), 3))
print('The bias is', bias, 'with significance', np.abs(significance))
```

```python
# estimating the variance weighted average of the fitted amplitude
estimated_amp = np.average(amps, axis=0, weights=1/np.array(vari))
estimated_var = np.average(vari, axis=0, weights=1/np.array(vari))
bias = estimated_amp - 1  # 1 is the originial true amplitude of the signal
significance = bias / np.mean(noise_chunk)
print('Amp =', round(estimated_amp, 3), 'with uncertainty', round(np.sqrt(estimated_var), 3))
print('The bias is', bias, 'with significance', np.abs(significance))
```
